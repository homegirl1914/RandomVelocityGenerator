```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
from velgen import flat_generator, cosine_fold_generator, linear_fault_generator, gaussian_salt_generator
```

```python
nx=100
ny=100
shape=(nx,ny)
```

```python
velseed5=[1.5, 2.0, 2.5, 3.0, 4.0]
```

```python
def test_generator(generator,nvel=3):
    plt.figure(figsize=[10,3])
    for i in range(nvel):
        v = generator.velocity()
        plt.subplot(1,nvel,i+1)
        plt.imshow(v.T)
```

# Flat

```python
generator = flat_generator(shape,velseed5)
test_generator(generator)
```

# Fold

```python
generator = cosine_fold_generator(shape,velseed5)
test_generator(generator)
```

# Fault

```python
generator = linear_fault_generator(shape,velseed5,max_nfaults=2)
test_generator(generator)
```

# Salt

```python
generator = gaussian_salt_generator(shape,velseed5,vel_salt=5.0)
test_generator(generator)
```
